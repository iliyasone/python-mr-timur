# Декоратор -- это функция которая принимает на вход другую функцию и возвращает третью

```python
import time

def decorator_no_args(func):
    def wrapper(ngsjgsd, bsdbgsdg):
        current = time.time()
        print(f'func {func.__name__} call')
        func(ngsjgsd, bsdbgsdg)
        print(f'func {func.__name__} call end by {time.time() - current}')

    return wrapper
```

```python
@decorator_no_args
def hello_world(a, b):
    time.sleep(5)
    print("Hello to this b**** world", a, b)
# hello_world = decorator(hello_world)
```

```python
hello_world(1, 22)
```

```python
hello_world()
```

```python
from functools import cache

@cache
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
@cache
def give_square(n: int):
    time.sleep(3)
    print(f'{n} and {n**2}')
    return n**2
```

```python
give_square(5)
```

```python
history = [0, 1] + [None] * 500

def fib(n):
    if history[n] is not None:
        return history[n]

    if n == 0:
        return 0
    if n == 1:
        return 1
    
    result = fib(n-1) + fib(n-2)
    
    history[n] = result
    return result

fib(100)
```

# Longest Common Subsequence

```python
p = "BATD"
q = "ABTACDZBDSD,V3T1B,XCB1-AS,,BZAXKFAS[T  YTF,XTVZMVBADBZBZST]"
 
def lcs(i, j) -> str:
    if i >= len(p):
        return ''
    if j >= len(q):
        return ''
    if p[i] == q[j]:
        return p[i] + lcs(i+1, j+1)
    s1 = lcs(i, j+1)
    s2 = lcs(i+1, j)

    if len(s1) > len(s2):
        return s1
    return s2

lcs(0, 0)
```
