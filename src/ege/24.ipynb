# 24_ege2021.txt

```python
with open('24-files/24_ege2021.txt') as file:
    text = file.read()

def solution(text: str) -> int:
    c = 1
    mx = 0
    for i in range(1, len(text)):
        if text[i-1] == 'a' and text[i] == 'd' \
        or text[i-1] == 'd' and text[i] == 'a':
            mx = max(mx, c)
            c = 1
        else:
            c += 1
    return max(mx, c)

assert solution('xyzxyz') == 6
assert solution('ad'+'x'*10) == 11
assert solution('ad'+'x'*10 + 'da') == 12
assert solution('ad'+'3'*3+'da'+'5'*5) == 6
assert solution('ad'+'3'*3+'da'+'5'*5 + 'ad') == 7


print(solution(text))
```

## Основная волна 2024

файл: https://inf-ege.sdamgia.ru/get_file?id=181109

```python
with open('files/24.69902.txt') as file:
    text = file.read()
# 874153

print(text.count('DE'), len(text))
def solution(text: str, window = 30) -> int:
    parts = text.replace('DE', 'D E').split()
    mx = 0
    for i in range(len(parts)):
        part = ''.join(parts[i:i+window+1])
        current_substring_len = len(part)

        if current_substring_len > mx:
            mx = current_substring_len
    return mx



print(solution('DEBBDECCCDEFFFDEG', window=2))
assert solution('DEBBDECCCDEFFFDEG', window=2) == 14
print(solution('DEDEDEDE', window=3))
assert solution('DEDEDEDE', window=3) == 7
assert solution('DDEDEDEDE', window=3) == 8
assert solution('DDEDE__DEDE', window=3) == 10

solution(text)
```

## Основная волна 2023 

файл: https://inf-ege.sdamgia.ru/get_file?id=137607

```python
with open('files/1_24.txt') as file:
    text = file.read()

def solution(text: str, window=150):
    indicies = []
    for i in range(len(text)-1):
        if text[i] =='T' and text[i+1] == 'T':
            indicies.append(i)
    mn = float('inf')
    for i in range(len(indicies) - window + 1):
        current_substring_len = len(text[indicies[i]:indicies[i+window-1]+2])
        if current_substring_len < mn:
            mn = current_substring_len
    return mn

assert solution('AATTAAATTAA', window=2) == 7
assert solution('AATTAAATTAATTT', window=2) == 3
solution(text)
```
