```python
f = open('files/26/26_8512.txt')
places = int(f.readline())
people = int(f.readline())
counter = 0
last_mesta = 0

bombardiro_crocodilo = []

for line in f.readlines():
    start, stop = map(int, line.split())
    bombardiro_crocodilo.append([start, stop])
    
bombardiro_crocodilo.sort()

cells = [0] * places

for start, stop in bombardiro_crocodilo:
    for i in range(len(cells)):
        if start > cells[i]:
            cells[i] = stop
            counter += 1
            last_mesta = i + 1
            break


print(counter, last_mesta)
f.close()
```

```python
class Open:
    def __init__(self, path):
        self.file = open(path)

    def __enter__(self):
        return self.file
    
    def __exit__(self, *args):
        self.close()

    def close(self):
        self.file.close()
```
